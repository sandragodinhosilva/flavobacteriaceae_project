## Metadata

```python
metadata = pd.read_csv("All_info.csv")
metadata = metadata[["Assembly accession","Genus","Family"]]
metadata.head(2)
```

## Virulence

```python
vir_orfs = pd.read_csv("Virulence_factors_counts2.csv")
vir_orfs = vir_orfs.rename(columns={"Unnamed: 0":"VFs"}).set_index("VFs").T
vir_orfs["Virulence orfs total"] = vir_orfs.sum(axis=1)
vir_orfs2 = vir_orfs.copy()
vir_orfs = vir_orfs.reset_index()
vir_orfs.head()
```

```python
vir_orfs3 = vir_orfs2.copy()
vir_orfs3 = vir_orfs3.drop(columns=["Virulence orfs total"]).T
vir_orfs3["Total in dataset"] = vir_orfs3.sum(axis=1)
vir_orfs3 = vir_orfs3.sort_values("Total in dataset", ascending=False)
vir_orfs3.head()
```

# Map

```python
map_vfs = pd.read_csv("VF_database_entries2.csv")
map_vfs.head()
```

```python
map_vfs["VFcode"].value_counts()
```

```python
vir_orfs4 = pd.merge(vir_orfs3, map_vfs, how="left", left_index=True, right_on="VF")
vir_orfs4.head(2)
```

```python
vir_orfs5 = vir_orfs4.groupby("name&vfcode").sum()#.drop(columns=["Total in dataset","VF","code","desc","VFcode","Organism"]).T
#vir_orfs5["Virulence orfs total"] = vir_orfs5.sum(axis=1)
vir_orfs5 = vir_orfs5.T.reset_index()
vir_orfs5.head(2)
```

### Merge with metadata:

```python
df = pd.merge(vir_orfs, metadata, left_on="index", left_index=True, right_on="Assembly accession")
df.head(2)
```

```python
df_namevfcode = pd.merge(vir_orfs5, metadata, left_on="index", left_index=True, right_on="Assembly accession")
df_namevfcode.head()
```

```python
df_namevfcode.to_csv("Virulence_counts_ready.csv", index=False)
```

# Plot

```python
df1 = df.groupby(['Genus']).filter(lambda x:x['Genus'].count()>7).groupby(['Genus', "Family"]).mean()
df1 = df1.reset_index()

df1 = df1.sort_values(by=["Family", "Genus"])
df1 = df1.drop(columns="Family")
df1 = df1.sort_index(axis=1, ascending=False)
df1.head()
```

```python
df2 = df1[["Genus","Virulence orfs total"]]
df2 = df2.set_index("Genus")
df2.head()
```

```python
df2.sort_values(by="Virulence orfs total", ascending=False).head(10)
```

```python
ax = df2.plot(kind="bar", figsize=(15,5))

plt.ylabel('Mean Virulence ORFs per genome')
plt.xticks(rotation=45, horizontalalignment='right')

plt.tight_layout()
plt.show()

ax.figure.savefig("Virulence_counts.svg", dpi=300)
```
