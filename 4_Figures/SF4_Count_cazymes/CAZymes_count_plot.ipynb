```python
mean = data.groupby(["Family"])["Nr_cazymes"].describe().round(1)
mean.head()
mean.loc["Flavobacteriaceae", "mean"]
```

# Plots

```python
final2 = final2.sort_values(by=["Family","Genus"])
#plt.figure(figsize=(20,8))
#sns.set_context("paper", rc={"font.size":15,"axes.titlesize":15,"axes.labelsize":15})   

def defaultPlotting(): 
    sns.set(rc={'figure.figsize':(11.7,8.27),"font.size":22,"axes.titlesize":24,"axes.labelsize":20},style="white")
defaultPlotting()

g = sns.catplot(
    data=final2, kind="bar", 
    x='Genus', hue="Family",dodge=False, legend_out=False,
    y='mean', 
    height=6, # make the plot 7 units high
    aspect=3) # height is n times width

g.despine(left=True)

#ax.add_artist(leg1)

g.map(plt.errorbar, "Genus", "mean", "std", ls='none', capsize=2, color="dimgrey")

#lines with means per Family
g.map(plt.axhline, y=mean.loc["Flavobacteriaceae", "mean"], ls='--', c='yellow')
g.map(plt.axhline, y=mean.loc["Weeksellaceae", "mean"], ls='--', c='purple')

#min and max points
sns.stripplot(x='Genus',y='max', data=final2, jitter=False, color="orangered", alpha=0.9)
sns.stripplot(x='Genus',y='min', data=final2, jitter=False, color="darkturquoise", alpha=0.9)

#labels
plt.xlabel('Genomes grouped per Genus', fontsize=15)
plt.ylabel('Average nr. CAZymes', fontsize=15)
#plt.title("CAZymes", fontsize=20)

for ax in g.axes.ravel():
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

import matplotlib.lines as mlines
red_line = mlines.Line2D([], [], color='orangered', marker='o', ls="",
                          markersize=7, label='Maximum count')
blue_line = mlines.Line2D([], [], color='royalblue', marker='o',ls="",
                          markersize=7, label='Minimum count')
black_line=mlines.Line2D([], [], color='dimgrey', #marker='',
                          markersize=7, label='Standard deviation')
yellow_line=mlines.Line2D([], [], color='yellow', #marker='',
                          markersize=7, label='Flavobacteriaceae mean     ')
purple_line=mlines.Line2D([], [], color='purple', #marker='',
                          markersize=7, label='Weeksellaceae mean')

leg1 = plt.legend(loc='upper left',borderaxespad=0.)
leg2 = plt.legend(handles=[red_line,blue_line, black_line], loc="upper right", borderaxespad=0.)#;bbox_to_anchor=(0, 0),

ax.add_artist(leg1)
ax.add_artist(leg2)

plt.legend(handles=[yellow_line, purple_line], loc="upper right", borderaxespad=0, bbox_to_anchor=(0.9, 1))

#ax = plt.gca()
#plt.tight_layout()

#plt.savefig("cazymes.png", tight=None)
plt.show()
```
